## Q1: We have been asked to check whether mean of the price is $800,000 or higher, hence we should apply the Z test 

## H0: Mean is 800,000
H1: Mean is greater than 800,000 

```python
property_df['Price'].mean()
```

```python
x_bar = 800000 #Sample Mean
mu = property_df['Price'].mean() #Population Mean
sigma = 6393.1 #Sample Standard Deviation
n = 13580 #Number of samples
```

```python
z = (x_bar - mu)/(sigma/math.sqrt(n))
```

```python
z
```

```python
p = (1 - stats.norm.cdf(z))
```

```python
p
```

```python
alpha=0.05 # Since significance level is said to be 5%
```

```python
if p <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

## This means that the mean is not $800,000 - the mean is greater than $800,000. 

## Q2. We need to check if the mean of prices of Summer 2016 = mean of prices of Winter 2016- hence we should apply the T test for independent samples- because the prices of Summer and WInter are independent of each other. 

## H0: Mean of Summer 2016=Mean of Winter 2016
## H1: Mean of Summer 2016 does not equal Mean of Winter 2016

```python
property_df.head()
```

```python
# Convert Date column to datetime (format m/dd/yyyy)
property_df["Date"] = pd.to_datetime(property_df["Date"], format="%d/%m/%Y")
```

```python
# Define date ranges for winter
range1_start = "2016-01-01"
range1_end   = "2016-03-31"

range2_start = "2016-10-01"
range2_end   = "2016-12-31"
```

```python
winter = (
    (property_df["Date"].between(range1_start, range1_end)) |
    (property_df["Date"].between(range2_start, range2_end))
)
```

```python
winter_prices = property_df.loc[winter, "Price"].to_numpy()
```

```python
# Define date ranges for summer
range3_start = "2016-04-01"
range3_end   = "2016-09-30"
```

```python
summer = (
    property_df["Date"].between(range3_start, range3_end) 
)
```

```python
summer_prices=property_df.loc[summer, "Price"].to_numpy()
```

```python
t_stat,p_value = stats.ttest_ind(winter_prices,summer_prices)
```

```python
t_stat
```

```python
p_value
```

```python
if p_value <= 0.05:
    print("Reject H0: Means average summer and average winter prices are different")
else:
    print("Fail to reject H0: Means average summer price and average winter price are equal")
```
